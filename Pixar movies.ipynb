```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
pixar_movies = pd.read_csv('PixarMovies.csv')
```

```python
pixar_movies.describe()
```

```python
pixar_movies.dtypes
```

```python
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str.rstrip('%').astype(float)
pixar_movies['International %'] = pixar_movies['International %'].str.rstrip('%').astype(float)
```

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score']*10
```

```python
filtered_pixar = pixar_movies.dropna(axis = 0)
```

```python
pixar_movies.set_index('Movie',inplace = True)
filtered_pixar.set_index('Movie',inplace = True)
```

```python
critic_reviews = pixar_movies[['RT Score','IMDB Score','Metacritic Score']]
```

```python
critic_reviews.plot(figsize = (10,6))
```

```python
critic_reviews.plot(kind = 'box',figsize = (9,5))
```

Rotten tomatoes have the highest scores, while metacritic has the widest range


```python
revenue_proportions = filtered_pixar[['Domestic %','International %']]
revenue_proportions.plot.bar()
```

```python
revenue_proportions.plot(kind = 'bar',stacked = True)
```

We can see as time goes there's a general increase in international views
